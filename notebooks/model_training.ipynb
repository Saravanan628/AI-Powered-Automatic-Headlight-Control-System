```python
import cv2
import time

# Open the laptop webcam (0 = default camera)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Set image count
img_count = 0

while True:
    ret, frame = cap.read()  # Capture frame
    if not ret:
        print("Failed to capture image.")
        break

    # Show the captured image in a window
    cv2.imshow("Webcam Capture", frame)

    # Wait for user input ('s' to save, 'q' to quit)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('s'):  
        img_name = f"image_{img_count}.jpg"
        cv2.imwrite(img_name, frame)  # Save image
        print(f"Image {img_name} saved!")
        img_count += 1  # Increase count for next image

    elif key == ord('q'):
        break  # Quit if 'q' is pressed

# Release the webcam
cap.release()
cv2.destroyAllWindows()

```
